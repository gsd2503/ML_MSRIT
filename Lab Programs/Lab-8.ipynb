```python
p_disease = 0.001
p_positive_given_disease = 0.98
p_negative_given_nodisease = 0.90
```

```python
p_nodisease = 1 - p_disease
p_positive_given_nodisease = 1 - p_negative_given_nodisease
```

```python
p_disease_given_positive = (p_positive_given_disease * p_disease) / ((p_positive_given_disease * p_disease) + (p_positive_given_nodisease * p_nodisease))
```

```python
print(f'Probability of having the disease given when the test result is positive {p_disease_given_positive:.2%}')
```
